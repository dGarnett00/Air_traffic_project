```python
#bar plot: Top 5 airlines by total passenger count
# Group by airline and sum passenger counts
top_airlines = df1.groupby('Operating Airline', as_index=False)['Passenger Count'].sum()
top_airlines = top_airlines.sort_values('Passenger Count', ascending=False).head(5)

plt.figure(figsize=(8, 5))
bar = sns.barplot(data=top_airlines, x='Operating Airline', y='Passenger Count', palette='Blues_d')

plt.title('Top 5 Airlines by Total Passenger Count')
plt.xlabel('Operating Airline')
plt.ylabel('Passenger Count')
plt.xticks(rotation=30, ha='right')
bar.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1_000_000:.1f}M' if x >= 1_000_000 else f'{x/1_000:.1f}K'))

plt.tight_layout()
plt.show()
```

```python
# Pie chart: Proportion of top 5 airlines by # of landings
top_landings = df2.groupby('Operating Airline', as_index=False)['Landing Count'].sum()
top_landings = top_landings.sort_values('Landing Count', ascending=False).head(5)

plt.figure(figsize=(6, 6))
plt.pie(
    top_landings['Landing Count'],
    labels=top_landings['Operating Airline'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('mako', n_colors=5)
)
plt.title('top 5 airlines by # of landings')    
plt.tight_layout()
plt.show()
```
