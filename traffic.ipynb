```python
# Bar Plot: Top 10 Airlines by Total Passenger Count (with Readable Labels)
"""
This bar plot visualizes the total number of passengers carried by the top 10 airlines in the dataset. The y-axis is formatted to show numbers in millions (M) or thousands (K) for clarity. Each bar is annotated with the exact value in a readable format.
"""
import matplotlib.ticker as mticker
top10_airlines = df1.groupby('Operating Airline', as_index=False)['Passenger Count'].sum().sort_values('Passenger Count', ascending=False).head(10)
plt.figure(figsize=(16,8))
barplot = sns.barplot(x='Operating Airline', y='Passenger Count', data=top10_airlines, hue='Operating Airline', palette='viridis', legend=False)
plt.title('Top 10 Airlines by Total Passenger Count', fontsize=18, fontweight='bold')
plt.ylabel('Total Passenger Count', fontsize=14)
plt.xlabel('Operating Airline', fontsize=14)
plt.xticks(rotation=35, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Format y-axis labels to show K/M
def human_format(num):
    if num >= 1e6:
        return f'{num/1e6:.1f}M'
    elif num >= 1e3:
        return f'{num/1e3:.1f}K'
    else:
        return str(int(num))
barplot.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: human_format(x)))
# Annotate each bar with the passenger count value in readable format
for i, row in top10_airlines.iterrows():
    barplot.text(i, row['Passenger Count'] + top10_airlines['Passenger Count'].max()*0.01, human_format(row['Passenger Count']), ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')
plt.subplots_adjust(left=0.08, right=0.98, bottom=0.18, top=0.92)
plt.show()
```

```python
# Pie chart: Proportion of flights by airline (top 5)
top_airlines = df1['Operating Airline'].value_counts().nlargest(5)
plt.figure(figsize=(6,6))
top_airlines.plot.pie(autopct='%1.1f%%')
plt.title('Top 5 Airlines by Number of Records')
plt.ylabel('')
plt.show()
```

```python
# Line chart: Passenger count over time (if you have a date column)
if 'Activity Period' in df1.columns:
    df1['Activity Period'] = pd.to_datetime(df1['Activity Period'], format='%Y%m')
    df1.groupby('Activity Period')['Passenger Count'].sum().plot(figsize=(12,6))
    plt.title('Passenger Count Over Time')
    plt.ylabel('Passenger Count')
    plt.xlabel('Date')
    plt.show()
```
